Concatination and JOIN

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3'], 
                    'C': ['C0', 'C1', 'C2', 'C3'], 
                    'D': ['D0', 'D1', 'D2', 'D3']}, 
                  index=[0,1,2,3])
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'], 
                    'B': ['B4', 'B5', 'B6', 'B7'], 
                    'C': ['C4', 'C5', 'C6', 'C7'], 
                    'D': ['D4', 'D5', 'D6', 'D7']}, 
                  index=[4,5,6,7])
```

```python
pd.concat([df1,df2])
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3'], 
                    'C': ['C0', 'C1', 'C2', 'C3'], 
                    'D': ['D0', 'D1', 'D2', 'D3']}, )


df2 = pd.DataFrame({'E': ['A4', 'A5', 'A6', 'A7'], 
                    'F': ['B4', 'B5', 'B6', 'B7'], 
                    'C': ['C0', 'C3', 'C6', 'C7'], 
                    'G': ['D4', 'D5', 'D6', 'D7']}, )
```

```python
pd.merge(df1, df2) # It merges df1 and df2 based on common columns
```

```python
pd.merge(df1, df2, on='C') # merge df1 and df2 based on column 'C'
```

```python
pd.merge(df1, df2, on='C', how='inner')
```

```python
pd.merge(df1, df2, on='C', how='outer') 
# take all values from df1 and df2, and fill missing values with NaN
```

```python
pd.merge(df1, df2, on='C', how='left')
# take all rows from df1 and the matching rows from df2
```

```python
pd.merge(df1, df2, on='C', how='right')
# take all rows from df2 and the matching rows from df1
```

#### `loc vs iloc`
- `loc` gets rows or columns with particular label.
- `iloc` gets rows or columns with particular integer position.

```python
s = pd.Series(list("abcdef"), index=[49, 48, 47, 0, 1, 2]) 
s
```

```python
s.loc[0]    # value at index label 0
```

```python
s.iloc[0]   # value at index location 0
```

```python
s.loc[0:1]  # rows at index labels between 0 and 1 (inclusive)
```
