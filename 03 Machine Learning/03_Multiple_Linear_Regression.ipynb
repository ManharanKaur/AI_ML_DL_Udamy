```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv('50_Startups.csv')
```

```python
df.head()
```

# `Divide the data into dependents vs independents`

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
X
```

```python
y
```

# `Feature Encoding - OneHotEncoder`

```python
from sklearn.compose import ColumnTransformer 
# ColumnTransformer apply different preprocessing steps to different columns in your dataset â€” all in one go.
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
X
```

# `Train Test Split`

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
X_test
```

```python
y_test
```

# `Feature Scaling`

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

# `Create LR Model`

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Predictions
y_pred = regressor.predict(X_test)
```

```python
# Plotting of y_test vs y_pred
plt.plot(y_test, color = 'blue', label = 'test')
plt.plot(y_pred, color = 'red', label = 'predictions')
plt.show()
```
